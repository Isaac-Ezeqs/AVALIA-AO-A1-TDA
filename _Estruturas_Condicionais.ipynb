```python
#Estruturas Condicionais

idade = int(input("Digite sua idade: "))

if idade < 0:
    print("Idade inválida.")
elif idade < 12:
    print("Entrada permitida apenas acompanhado dos pais.")
elif idade < 18:
    print("Entrada permitida somente com autorização.")
else:
    print("Entrada liberada! Aproveite o evento.")
```
